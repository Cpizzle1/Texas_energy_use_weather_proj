```python
def make_date_time_col_city(df):
    
    df['Date']= df['Date'].astype(str)
    df['Hour Number'] = df['Hour Number'].astype(str)
    df['New_datetime'] = df['Date'].map(str) + " " + df['Hour Number']
    df['Hour Number'] = df['Hour Number'].astype(int)
    
    return df
```

```python
converted_houston_temp_24 =  make_date_time_col_city(houston_temp_24)
```

```python
converted_houston_temp_24
```

```python
converted_houston_temp_24['New_datetime']= converted_houston_temp_24['New_datetime'].apply(lambda x:f'{x}:00:00')
```

```python
converted_houston_temp_24.info()
```

```python
converted_houston_temp_24['New_datetime'] = pd.to_datetime(converted_houston_temp_24['New_datetime'],infer_datetime_format=True, format ='%m/%d/%Y %H')
```
